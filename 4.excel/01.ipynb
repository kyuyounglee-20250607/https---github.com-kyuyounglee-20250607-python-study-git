```python
import pandas as pd
url = 'https://raw.githubusercontent.com/tsdata/pandas-data-analysis/refs/heads/main/part7/data/wolesale_customers.csv'
df = pd.read_csv(url)
df.head()
```

csv파일을 읽고 데이터 처리

```python
# row 데이터
# 데이터 프레임에서 iloc 인덱스로 처리
# .iloc[행, 열]  .iloc[2:4, 3:6]  .iloc[ [1,5], [5,7]   ]
df.iloc[ 0: , 0:5 ]
```

```python
df['total'] = df.iloc[:, 2:].sum(axis=1)
df
```

```python
# channel 거래 채널
# Regin 고객 지역
# 채널이 1 : 호텔/레스토랑/카페
# 2. 소매점
# 지역 1 : Lisbon  2: Oporto   3: 기타
# fresh 신선식품  Milk 유제품  Grocery 식료품  
# Frozen 냉동식품 Dtergents_Papaer 세제 및 종이류  Delicassen 고급식품

df.columns
```

```python
df.rename(columns={'Channel': '거래처' })
```

```python
names = ['채널','지역','신선식품','유제품','식료품','냉동식품','세제및종이류','고급식품']
dict(zip(df.columns,names))
```

```python
df = df.rename(columns=dict(zip(df.columns,names)))
df.head()
```

```python
df['채널'] = df['채널'].map( {1:"호델/레스토랑/카페", 2:'소매점'}  )
df['지역'] = df['지역'].map( {1:'lisbon', 2:'Oporto', 3:'기타'}  )
df.head()
```

```python
df.groupby('채널')['신선식품'].sum()
```

```python
df.groupby('채널')['신선식품'].agg( ('sum','mean') )
```

```python
df.groupby('채널').agg(신선식품_합계 = ('신선식품','sum'))
```

```python
by_channel = df.groupby('채널').agg(
    신선식품_합계 = ('신선식품','sum'),
    유제품_합계 = ('유제품','sum'),
    식료품_합계 = ('식료품','sum'),
    냉동식품_합계 = ('냉동식품','sum'),
    세제및종이류_합계 = ('세제및종이류','sum'),
    고급식품_합계 = ('고급식품','sum'),
    total_합계 = ('total','sum'),
    )
by_channel
```

```python
by_region =df.groupby('지역').agg(
    신선식품_합계 = ('신선식품','sum'),
    유제품_합계 = ('유제품','sum'),
    식료품_합계 = ('식료품','sum'),
    냉동식품_합계 = ('냉동식품','sum'),
    세제및종이류_합계 = ('세제및종이류','sum'),
    고급식품_합계 = ('고급식품','sum'),
    total_합계 = ('total','sum'),
    )
by_region
```

```python
# 2. 엑셀 파일로 저장
excel_file = 'test.xlsx'
writer = pd.ExcelWriter(excel_file, engine='xlsxwriter')
resion_df.to_excel(writer, sheet_name='데이터', index=False)

resion_df = resion_df.iloc[:-1]

# 3. 워크북 및 워크시트 객체 가져오기
workbook = writer.book
worksheet = writer.sheets['데이터']

# 4. 차트 생성
chart = workbook.add_chart({'type': 'column'})

# 5. 차트 데이터 계열 추가 (각 품목을 계열로)
num_regions = len(resion_df)
# Iterate over each product column to add it as a series
for i, col_name in enumerate(resion_df.columns[1:], 1):
    chart.add_series({
        'name':       ['데이터', 0, i],  # 품목 이름 (e.g., '신선식품_합계')
        'categories': ['데이터', 1, 0, num_regions, 0], # 지역 이름 (X-axis)
        'values':     ['데이터', 1, i, num_regions, i], # 해당 품목의 지역별 값
        'data_labels': {'value': True, 'num_format': '#_-'},
    })

# 6. 차트 축 및 제목 설정
chart.set_title({'name': '지역별 상품 판매 현황'})
chart.set_x_axis({'name': '지역'})
chart.set_y_axis({'name': '수량'})

# 7. 차트 스타일 및 크기 설정
chart.set_style(10)
worksheet.insert_chart('A10', chart, {'x_scale': 2, 'y_scale': 2})

# 8. 엑셀 파일 저장
writer.close()

print(f"엑셀 파일이 생성되었습니다: {excel_file}")
```

```python

import pandas as pd

# 1. 기존 엑셀 파일 읽기
input_excel_file = 'report.xlsx'
df = pd.read_excel(input_excel_file, sheet_name='채널')

# 2. 차트를 추가할 새로운 엑셀 파일 설정
output_excel_file = 'report_chart.xlsx'
writer = pd.ExcelWriter(output_excel_file, engine='xlsxwriter')
df.to_excel(writer, sheet_name='데이터', index=False)

# 3. 워크북 및 워크시트 객체 가져오기
workbook = writer.book
worksheet = writer.sheets['데이터']

# 4. 차트 생성
chart = workbook.add_chart({'type': 'column'})

# 5. 차트 데이터 계열 추가 (각 품목을 계열로)
num_rows = len(df)
for i, col_name in enumerate(df.columns[1:], 1):
    chart.add_series({
        'name':       ['데이터', 0, i],  # 품목 이름
        'categories': ['데이터', 1, 0, num_rows, 0], # 지역 이름 (X-axis)
        'values':     ['데이터', 1, i, num_rows, i], # 해당 품목의 지역별 값
        'data_labels': {'value': True, 'num_format': '#_-'},
    })

# 6. 차트 축 및 제목 설정
chart.set_title({'name': '지역별 상품 판매 현황'})
chart.set_x_axis({'name': '지역'})
chart.set_y_axis({'name': '수량'})

# 7. 차트 스타일 및 크기 설정
chart.set_style(10)
worksheet.insert_chart('A10', chart, {'x_scale': 2, 'y_scale': 2})

# 8. 엑셀 파일 저장
writer.close()

print(f"엑셀 파일이 생성되었습니다: {output_excel_file}")
```
